```python
# import the modules
import os
from os import listdir
data_file = []
data_img = []
ds  = []
# get the path/directory
folder_dir = r'C:\Users\z_ref_analyst1\Desktop\CCR_Img_regression\train'
for images in os.listdir(folder_dir):
       im = Image.open(images)
       # im_grey = im.convert('LA') 
       im_grey = im.convert('LA') #convert to grayscale. For the “P” mode, this method translates pixels through the palette.
       width, height = im.size

       total = 0
       for i in range(0, width):
            for j in range(0, height):
                total += im_grey.getpixel((i,j))[0]

       mean = total / (width * height)
       print(mean, images)
       data_file.append(mean)
       data_img.append(images)
       #data_file.append(stat.mean)
       #data_img.append(images)
       #data = print(stat.mean, images)
```

```python
data_file_fr = pd.DataFrame(data=data_file, columns=['Grayscale'])
data_file_fr.head()
```

```python
data_img_fr = pd.DataFrame(data = data_img, columns = ['files'])
data_img_fr.head()
```

```python
output_frame = pd.concat([data_file_fr, data_img_fr], axis = 1) # Creating Dataframe
output_frame.head()
```
